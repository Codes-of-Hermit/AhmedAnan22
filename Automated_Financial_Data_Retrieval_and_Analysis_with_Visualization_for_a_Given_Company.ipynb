```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def get_financial_data(company_name):
    # Fetch the stock ticker symbol
    company_info = yf.Ticker(company_name)

    # Get historical market data
    hist = company_info.history(period="10y")

    # Get financials
    financials = company_info.financials.T
    balance_sheet = company_info.balance_sheet.T
    cashflow = company_info.cashflow.T

    return hist, financials, balance_sheet, cashflow

def plot_stock_price(hist):
    plt.figure(figsize=(10, 5))
    plt.plot(hist['Close'], label='Stock Price')
    plt.title('Stock Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_financial_ratios(financials):
    ratios = pd.DataFrame()
    ratios['Revenue'] = financials['Total Revenue']
    ratios['Net Income'] = financials['Net Income']
    ratios['Operating Income'] = financials['Operating Income']
    ratios['Gross Profit'] = financials['Gross Profit']

    ratios.plot(kind='bar', figsize=(10, 5))
    plt.title('Financial Ratios Over Time')
    plt.xlabel('Year')
    plt.ylabel('Amount')
    plt.grid(True)
    plt.show()

def financial_analysis(company_name):
    # Fetch financial data
    hist, financials, balance_sheet, cashflow = get_financial_data(company_name)

    # Plot stock price
    plot_stock_price(hist)

    # Plot financial ratios
    plot_financial_ratios(financials)

if __name__ == "__main__":
    company_name = input("Enter the company's ticker symbol (e.g., AAPL for Apple): ")
    financial_analysis(company_name)
```
